```python
def btf(x):
    btf=engine.below_the_fold(x)
    try:
        totalAmenities.append(btf['payload']['amenitiesInfo']['totalAmenities'])
    except:
        totalAmenities.append(None)
    try:
        beds.append(btf['payload']['publicRecordsInfo']['basicInfo']['beds'])
    except:
        beds.append(None)
    try:
        baths.append(btf['payload']['publicRecordsInfo']['basicInfo']['baths'])
    except:
        baths.append(None)
        print('baths'+str(x))
    try:
        propertyTypeName.append(btf['payload']['publicRecordsInfo']['basicInfo']['propertyTypeName'])
    except:
        propertyTypeName.append(None)
    try:
        numStories.append(btf['payload']['publicRecordsInfo']['basicInfo']['numStories'])
    except:
        numStories.append(None)
        print(x)
    try:
        yearBuilt.append(btf['payload']['publicRecordsInfo']['basicInfo']['yearBuilt'])
    except:
        yearBuilt.append(None)
        print('yearBuilt'+str(x))
    try:
        yearRenovated.append(btf['payload']['publicRecordsInfo']['basicInfo']['yearRenovated'])
    except:
        yearRenovated.append(None)
        print(x)
    try:
        totalSqFt.append(btf['payload']['publicRecordsInfo']['basicInfo']['totalSqFt'])
    except:
        totalSqFt.append(None)
        print('totalSqFt'+str(x))
    try:
        lotSqFt.append(btf['payload']['publicRecordsInfo']['basicInfo']['lotSqFt'])
    except:
        lotSqFt.append(None)
        print('lotSqFt'+str(x))
    try:
        taxesDue.append(btf['payload']['publicRecordsInfo']['taxInfo']['taxesDue'])
    except:
        taxesDue.append(None)
        print('taxesDue'+str(x))
    try:
        eventDescription.append(btf['payload']['propertyHistoryInfo']['events'][0]['eventDescription'])
    except:
        eventDescription.append(None)
    try:
        eventDateString.append(btf['payload']['propertyHistoryInfo']['events'][0]['eventDateString'])
    except:
        eventDateString.append(None)
        print(x)
    try:
        event=btf['payload']['propertyHistoryInfo']['events']
        try:
            count=0
            for e in event:
                try:
                    price.append(e['price'])
                    break
                except:
                    count+=1
                    if count==len(event):
                        price.append(None)
                    else:
                        continue
        except:
            price.append(None)
        try:
            for e in event:
                if eventDescription == 'Sold (Public Records)':
                    if e['eventDescription'] == 'Sold (Public Records)':
                        sold_date=e['eventDate']
                    if e['eventDescription'] == 'Listed':
                        listed_date=e['eventDate']
                        days_on_market=(sold_date-listed_date)/86400000
                        break
                else:
                    days_on_market=None
            daysOnMarket.append(days_on_market)
        except:
            daysOnMarket.append(None)
    except:
        price.append(None)
        daysOnMarket.append(None)
    try:
        countyName.append(btf['payload']['publicRecordsInfo']['countyName'])
    except:
        countyName.append(None)
    return totalAmenities, beds, baths, propertyTypeName, numStories, yearBuilt, yearRenovated, totalSqFt, lotSqFt, taxesDue, eventDescription, eventDateString, daysOnMarket, price, countyName
```

```python
def publicSchools(x):
    try:
        btf=engine.below_the_fold(x)
        sth=btf['payload']['schoolsAndDistrictsInfo']['servingThisHomeSchools']
        elementary_rating=0
        middle_rating=0
        high_rating=0
        elementary_ratio=0
        middle_ratio=0
        high_ratio=0
        elementary_student_ct=0
        middle_student_ct=0
        high_student_ct=0
        esc=0
        msc=0
        hsc=0
        for i in sth:
            name=i['name']
            elementary=i['elementary']
            middle=i['middle']
            high=i['high']
            if "Learning" in name:
                continue
            if elementary == True:
                try:
                    elementary_rating=elementary_rating+(((11-i['greatSchoolsRating'])+(11-i['parentRating']))/2*float(i['distanceInMiles']))
                    elementary_ratio=elementary_ratio+i['studentToTeacherRatio']
                    elementary_student_ct=elementary_student_ct+i['numberOfStudents']
                    esc=esc+1
                except:
                    continue
            if middle == True:
                try:
                    middle_rating=middle_rating+(((11-i['greatSchoolsRating'])+(11-i['parentRating']))/2*float(i['distanceInMiles']))
                    middle_ratio=middle_ratio+i['studentToTeacherRatio']
                    middle_student_ct=middle_student_ct+i['numberOfStudents']
                    msc=msc+1
                except:
                    continue
            if high == True:
                try:
                    high_rating=high_rating+(((11-i['greatSchoolsRating'])+(11-i['parentRating']))/2*float(i['distanceInMiles']))
                    high_ratio=high_ratio+i['studentToTeacherRatio']
                    high_student_ct=high_student_ct+i['numberOfStudents']
                    hsc=hsc+1
                except:
                    continue
        try:
            publicStudentToTeacherRatio=(elementary_ratio/max(1,esc)+middle_ratio/max(1,msc)+high_ratio/max(1,hsc))/3
            publicStudentTeacherRatio.append(publicStudentToTeacherRatio)
        except:
            publicStudentTeacherRatio.append(None)
        try:
            publicRating=(elementary_rating/max(1,esc)+middle_rating/max(1,msc)+high_rating/max(1,hsc))/3
            publicSchoolRate.append(publicRating)
        except:
            publicSchoolRate.append(None)
        try:
            publicSchoolCnt=(elementary_student_ct/max(1,esc)+middle_student_ct/max(1,msc)+high_student_ct/max(1,hsc))/3
            publicStudentCnt.append(publicSchoolCnt)
        except:
            publicStudentCnt.append(None)
    except:
        publicStudentTeacherRatio.append(None)
        publicSchoolRate.append(None)
        publicStudentCnt.append(None)
    return publicStudentTeacherRatio, publicSchoolRate, publicStudentCnt
```

```python
def amenities(x):
    try:
        btf=engine.below_the_fold(x)
        sg=btf['payload']['amenitiesInfo']['superGroups']
        amenitiesList={}
        ct=0
        for s in sg:
            ags=s['amenityGroups']
            for ag in ags:
                aes=ag['amenityEntries']
                for ae in aes:
                    key=ae['referenceName']
                    value=ae['amenityValues'][0]
                    amenitiesList[key]=value
        amenityList.append(amenitiesList)
    except:
        amenityList.append(None)
    return amenityList
```

```python
addresses=[]
for i in range(0,len(cities)):
    currCityNum= str(city_numbers[i])
    currCity=str(cities[i])
    req_headers = {
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'en-US,en;q=0.8',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
    }
    s=requests.Session()
    url = 'https://www.redfin.com/city/'+currCityNum+'/OR/'+currCity
    print(url)
    r = s.get(url, headers=req_headers)
    print(r.status_code)
    soup=BeautifulSoup(r.content, "html.parser")
    print(cities[i])
    sleep(2)
    addressesOnPage= soup.find_all('div', {'class': 'link-and-anchor'})
    for i in addressesOnPage:
        addresses.append(i.text)
    try:
        pages=int(soup.find_all('a', {'class': 'clickable goToPage'})[-1].text)
        print(pages)
        for page in range(2,pages+1):
            url = 'https://www.redfin.com/city/'+currCityNum+'/OR/'+currCity+'/page-'+str(page)
            print(url)
            r = s.get(url, headers=req_headers)
            soup=BeautifulSoup(r.content, "html.parser")
            sleep(2)
            addressesOnPage= soup.find_all('div', {'class': 'link-and-anchor'})
            for i in addressesOnPage:
                addresses.append(i.text)
    except:
        print('out of range')
```
